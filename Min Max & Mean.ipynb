## 1. LINE

In a line graph, the values of the variable are plotted as points on a graph and connected by a line. The x-axis typically represents time or another continuous variable, while the y-axis represents the values of the variable being measured.

We are picking data randomly for creating list and we gonna make two types of list which are x and y.

```python
x = [10, 20, 30, 40, 50, 60]
y = [2,4,6,8,10,12]
```

```python
plt.figure(figsize=(8,5))
plt.plot(x,y, color="red", label="data")
plt.xlabel("X axis", fontsize="24")
plt.ylabel("Y axis", fontsize="24")
plt.title("Line", fontsize="24")
plt.grid("grey", alpha=0.5)
plt.legend()
plt.show()
```

Now, we can see in above graph that data is linear,x is directly proportional to y, meaning that as the values of x increase, the values of y increase proportionally by a constant factor. The data points can be represented by the equation y = mx + b, where m represents the slope of the line and b represents the y-intercept.

```python
len(y)
```

```python
len(x)
```

In Next step, we are using the axhline feature in the matplotlib.pyplot library to plot a horizontal line on the graph. This line will be drawn at a specific y-value, which is specified by the y parameter. In this case, the line will be colored red and dashed to distinguish it from the other lines on the graph.

we will specify the horizontal extents of the line by setting the xmin and xmax parameters. These values determine the starting and ending x-coordinates of the line, respectively. The values are specified as fractions of the total width of the graph, with 0.05 representing the leftmost 5% of the graph and 0.95 representing the rightmost 5%.

```python
plt.figure(figsize=(8,5))
plt.plot(x, y, color="red", label="data")
#add a horizontal line across the plot 
plt.axhline(y=max(y), xmin=0.05, xmax=0.95, color='red', linestyle="--")
plt.text(20, 12.5, "High")
plt.axhline(y=min(y), xmin=0.05, xmax=0.95, color='red', linestyle="--")
plt.text(20, 2.5, "Low")
plt.xlabel("X axis", fontsize="24")
plt.ylabel("Y axis", fontsize="24")
plt.title("Line", fontsize="24")
plt.grid(color="grey", alpha=0.5)
plt.legend()
plt.show()
```

Declaring the function mean_color. It will take three parameters: x, y, and mean. We are using a for loop to iterate over x using the range and len functions. Similarly, we are iterating over the index values to match the condition that says, if the value at an index is greater or less than mean, append the corresponding color to a list data structure called colors. Finally, the function returns the colors list.

```python
def mean_color(x, y, mean):
    colors = []
    for i in range(len(x)):
        if y[i] > mean:
            colors.append("red")
        else:
            colors.append("blue")
    return colors
```

The plt.gca() function returns the current axes instance on the current figure. This is useful when you want to modify properties of the plot or add additional elements to it.

```python
#calling function
colors = mean_color(x, y, mean)
plt.figure(figsize=(8,5))
#get current axes
ax = plt.gca() 
#add a horizontal line across the plot 
plt.axhline(mean, color="black",linestyle="--")
plt.text(20,25, "mean line")
plt.bar(x, y, color=colors)
plt.show()
```

Similary, we are declaring color_fun function to generate four different color accoring to quadrant. we gonna consider mean and plot horizonatl and vertical line on mean. the logical and comparision is the game changer.

```python
def color_func(x, y,mean_x,mean_y):
    if x > mean_x and y > mean_y:
        return 'red'
    elif x > mean_x and y < mean_y:
        return 'blue'
    elif x < mean_x and y > mean_y:
        return 'green'
    else:
        return 'black'
```

```python
colors = [color_func(x[i], y[i], mean_x, mean_y) for i in range(len(x))]
```

## MEAN LINE FOR DISTRIBUTING INTO FOUR QUADRANT

```python
plt.figure(figsize=(8,5))
plt.scatter(x,y, color = colors)
plt.xlabel("X axis", fontsize="24")
plt.ylabel("Y axis", fontsize="24")
plt.axhline(50.5, color="red",linestyle="--")
plt.axvline(50.5, color="red",linestyle="--")
plt.title("Scatter", fontsize="24")
plt.show()
```
